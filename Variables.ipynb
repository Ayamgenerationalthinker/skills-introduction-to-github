Time Converter: Given a variable total_seconds (e.g., 5000), calculate how many full minutes and remaining seconds that is using floor division and modulo.

```python
total_seconds = 5000
minutes = total_seconds // 60
seconds = total_seconds % 60
print(minutes)
print(seconds)
```
