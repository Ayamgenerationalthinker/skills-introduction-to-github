ATM Simulation: Check if a withdrawal_amount is greater than the balance. If so, print "Insufficient funds"; otherwise, deduct the money.

```python
balance = input("What is your balance?")
withdrawal_amount = input("How much do you want to withdraw?")
new_balance = int(balance) - int(withdrawal_amount)

if int(withdrawal_amount) > int(balance):
  print("Insufficient funds")
else:
  print(f"Money deducted.  Your new balance is: {new_balance}" )
```

Grading System: Given a score, print 'A' (90+), 'B' (80-89), 'C' (70-79), or 'F' (below 70).

```python
score = input("Please enter your score")
if int(score) >= 90:
  print("A")
elif int(score) >= 80:
  print("B")
elif int(score) >= 70:
  print("C")
else:
  print("F")
```

Password Strength: Check if a password string has at least 8 characters. If it does, print "Strong"; otherwise, "Weak."

```python
password = input("Please enter your password")
if len(password) >= 8:
  print("Strong")
else:
  print("Weak")
```

Login Logic: Check if username == "admin" AND password == "1234". Print "Access Granted" or "Denied."

```python
username = input("Please enter your username: ")
password = input("Please enter your password: ")
if username == "admin"  and password == "1234":
  print("Access Granted")
else:
  print("Denied")
```

Leap Year Checker: Determine if a year is a leap year (divisible by 4, but not by 100 unless also divisible by 400).

```python
year = int(input("Please enter the year: "))
if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
  print("Leap year")
else:
  print("Not leap year")
```

Movie Theater Pricing: If age < 12, price is $5. If age > 65, price is $7. Everyone else pays $10.

```python
age = int(input("Please what is your age: "))
if age < 12:
  print("Your price is $5")
elif age > 65:
  print("Your price is $7")
else:
  print("Your price is $10")
```

FizzBuzz (Simple): If a number is divisible by 3, print "Fizz." If by 5, print "Buzz." If both, "FizzBuz

```python
number = int(input("Please enter your number "))
if number % 3 == 0 and number % 5 == 0:
  print("FizzBuzz")
elif number % 3 == 0:
  print("Fizz")
elif number % 5 == 0:
  print("Buzz")
```
